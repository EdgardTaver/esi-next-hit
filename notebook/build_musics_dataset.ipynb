# Build musics dataset

Jupyter notebook simples apenas para ter uma base de músicas.

Dados obtidos [aqui](https://zenodo.org/records/4778563).

Observação: dependências adicionais, além do `requirements.txt`, são necessárias para executar o código abaixo.

```python
import pandas as pd
```

```python
artists = pd.read_csv("spotify_artists_info_complete.csv", sep="\t")

display(artists.head())
display(artists.shape)
```

```python
artists = artists[artists['image_url'].notna()]

display(artists.head())
display(artists.shape)
```

```python
import ast
artists['genres'] = artists['genres'].apply(ast.literal_eval)
artists = artists[artists['genres'].apply(lambda x: len(x) > 0)]

display(artists.head())
display(artists.shape)
```

```python
artists['first_genre'] = artists['genres'].apply(lambda x: x[0] if len(x) > 0 else None)

display(artists.head())
display(artists.shape)
```

```python
hits = pd.read_csv("spotify_hits_dataset_complete.csv", sep="\t")

display(hits.head())
display(hits.shape)
```

```python
hits.isna().sum()
```

```python
hits.columns
```

```python
hits = hits[hits['song_name'].notna()]

display(hits.head())
display(hits.shape)
```

```python
import ast
hits['artist_name'] = hits['artist_name'].apply(ast.literal_eval)
hits['first_artist'] = hits['artist_name'].apply(lambda x: x[0] if len(x) > 0 else None)

display(hits.head())
display(hits.shape)
```

```python
hits["first_artist"].isna().sum()
```

```python
hits_cols = ["song_name", "first_artist"]
artist_cols = ["name", "image_url", "first_genre"]


merged = pd.merge(
    hits[hits_cols],
    artists[artist_cols],
    left_on="first_artist",
    right_on="name",
    how="inner")

merged.rename(columns={
    "name": "artist_name",
    "first_genre": "genre"
}, inplace=True)

merged.drop(columns=["first_artist"], inplace=True)

display(merged.head())
display(merged.shape)
```
